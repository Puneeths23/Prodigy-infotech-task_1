# DATA VISALIZATION

```python
plt.figure(figsize=(40, 20))
plt.bar(x=df['Countries and areas'], height=df['Youth_15_24_Literacy_Rate_Male'], width=0.5)
plt.title('Literacy Rate of Youth (Aged 15-24) - Male')
plt.xlabel('Countries and areas')
plt.ylabel('Youth Literacy Rate (Male)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
filtered_data = df[df['Youth_15_24_Literacy_Rate_Male'].notna() & (df['Youth_15_24_Literacy_Rate_Male'] != 0)]

# Plot the filtered data
plt.figure(figsize=(24, 10))
plt.bar(x=filtered_data['Countries and areas'], height=filtered_data['Youth_15_24_Literacy_Rate_Female'], width=0.5)
plt.title('Literacy Rate of Youth (Aged 15-24) - Female')
plt.xlabel('Countries and areas')
plt.ylabel('Youth_15_24_Literacy_Rate_Female')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
plt.figure(figsize=(40,20))
plt.bar(x=df['Countries and areas'], height=df['Youth_15_24_Literacy_Rate_Female'], width=0.5)
plt.title('Literacy Rate of Youth (Aged 15-24) - Female')
plt.xlabel('Countries and areas')
plt.ylabel('Youth Literacy Rate (Male)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
filtered_data = df[df['Youth_15_24_Literacy_Rate_Female'].notna() & (df['Youth_15_24_Literacy_Rate_Female'] != 0)]

# Plot the filtered data
plt.figure(figsize=(24, 10))
plt.bar(x=filtered_data['Countries and areas'], height=filtered_data['Youth_15_24_Literacy_Rate_Female'], width=0.5)
plt.title('Literacy Rate of Youth (Aged 15-24) - Female')
plt.xlabel('Countries and areas')
plt.ylabel('Youth_15_24_Literacy_Rate_Female')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
filtered_data = df['Youth_15_24_Literacy_Rate_Female'].dropna().replace(0, np.nan)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(filtered_data, bins=20, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Literacy Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Youth (Aged 15-24) Literacy Rate - Female')

# Show plot
plt.show()
```
